```python
import requests
import pandas as pd
import re
import numpy as np
from bs4 import BeautifulSoup
import datetime
```

```python
URL = 'http://www.hzmedia.com.cn/books11119011'
r = requests.get(url=URL)
WebText = BeautifulSoup(r.text,"lxml")
r.close()

ContentList = [each.replace(' ','').replace('\u3000',' ') for each in re.findall('<br/>\n(.*?)\n.*<br/>',WebText.prettify())]
ContentDf = pd.DataFrame({'目录':ContentList})
ContentDf.csv('')
```

```python
for page in range(1,21):
    URL = 'http://www.hzmedia.com.cn/w/action.ashx'
    r = requests.post(URL,{'m': 'e6f1f3df-b7e2-451e-905a-ccf391549b58_1','p': '%d'%(page)})
    WebText = BeautifulSoup(r.text,"lxml")
    r.close()
    ResultList=re.findall('>(\n.*?\n).*?<',WebText.prettify())
    print(ResultList)
```
