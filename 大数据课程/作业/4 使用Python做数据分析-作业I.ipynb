## 对于所有国家，每年人均二氧化碳的排放量在如何变化？

#### 绘制从1960年至2011的世界平均值图

载入matplotlib模块用于绘制图形

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

生成图

```python
mydf = df.copy()
```

```python
pd.to_datetime(mydf[mydf.columns[2:54]].mean().index)
```

```python
#Assign x,y values
world_x = pd.to_datetime(mydf[mydf.columns[2:54]].mean().index)
world_y = mydf[mydf.columns[2:54]].mean()

#Plot
plt.plot(world_x, world_y)
plt.title("CO2 Emissions per capita averaged across all countries")
plt.ylabel("Mean CO2 Emissions (metric tons/capita)")
plt.xlabel("Year")
```

## 2011年哪个地区的二氧化碳排放量（CO2 emissions）最高?

#### 一共有哪些地区?

```python
mydf.Region
```

#### 返回唯一的地区（去掉重名地区）:

```python
mydf.Region.unique()
```

#### 根据地区分组，然后使用mean函数做聚合

```python
by_region = mydf.groupby('Region').agg('mean') #groupby then aggregate
by_region['2011']
```

#### 获得聚合后的结果

```python
by_region['2011']
```

#### 绘制2011年地区平均二氧化碳排放量图

```python
plt.bar(range(len(by_region['2011'])),by_region['2011'].values)
plt.xticks(range(len(by_region['2011'])), by_region['2011'].index, rotation=70)
plt.ylabel('CO2 Emission (metric ton/capita)')
plt.title('Mean CO2 Emission by Region in 2011')
```

## 2011年哪个收入群体有最高的平均二氧化碳排放量?

第一，先在mydf中得到收入群体种类，注意使用**unique**去除重复。

```python
## MY CODE BELOW
mydf.IncomeGroup.unique()

```

然后，对数据帧进行分组，通过IncomeGroup这一列进行分组，然后聚合求平均

```python
## MY CODE BELOW

# Remember to limit to 2011. 
# Assign result to a new variable, by_incomegroup

by_IncomeGroup = mydf.groupby('IncomeGroup').agg('mean')
by_IncomeGroup['2011']
```

在条线图(bar graph)中绘制结果

```python
## MY CODE BELOW

plt.bar(range(len(by_IncomeGroup['2011'])),by_IncomeGroup['2011'].values)
plt.xticks(range(len(by_IncomeGroup['2011'])), by_IncomeGroup['2011'].index, rotation=70)
plt.ylabel('CO2 Emission (metric ton/capita)')
plt.title('Mean CO2 Emission by IncomeGroup in 2011')
```

## 展示1960年至2011年的二氧化碳排放量趋势

我们可以创建一个折线图来展示这个趋势

```python
def plot_countries(mydf):
    from pylab import rcParams
    rcParams['figure.figsize'] = (12,6)
    import numpy as np

    colormap = plt.cm.gist_ncar
#     plt.gca().set_color_cycle([colormap(i) for i in np.linspace(0, 0.9, len(range(1960,2012)))])
    plt.gca().set_prop_cycle('color',plt.cm.Spectral(np.linspace(0, 0.9, len(range(1960,2012)))))

    for i in range(0,len(mydf)):
        plt.plot(mydf.iloc[i,][2:54])
    plt.xticks(range(len(range(1960,2012))),mydf.columns.values[2:54], rotation = 70)
    #plt.legend(mydf["Country Name"].values, loc='upper left')
    plt.legend(mydf["Country Name"].values, ncol=4, loc='upper center', 
               bbox_to_anchor=[0.4, -0.1], 
               columnspacing=1.0, labelspacing=0.0,
               handletextpad=0.0, handlelength=1.0,
               fancybox=True)
    plt.ylabel("CO2 Emission (metric ton/capita)")
    plt.title("CO2 Emission Per Capita by Country and Year")
    plt.show()
```

```python
plot_countries(mydf)
```
