## 简单的散点图

另一个常用的绘图类型是简单的散点图，这是线图的近亲。这里的点用点、圆或其他形状分别表示，而不是用线段连接。我们将首先设置notebook，以绘制和导入我们将使用的函数:

````python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
````

## 用``plt.plot``来绘制散点图
在上一个实验中我们学习了用``plt.plot``/``ax.plot``来绘制线形图，同样的函数也可以产生散点图。

````python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black');
````

函数调用中的第三个参数是一个字符，它表示用于绘图的符号类型。正如您可以指定诸如``'-'``, ``'--'`` 这样的选项来控制线条样式，标记风格也有自己的一套简短的字符串代码。可用符号的完整列表可以在``plt.plot``的文档中看到。或者在Matplotlib的在线文档中。大多数可能性都是相当直观的，我们会在这里展示一些更常见的可能性:

````python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
````

为了获得更多的可能性，这些字符代码可以与线和颜色代码一起使用，并与连接它们的线一起使用:

````python
plt.plot(x, y, '-ok');
````

额外的关键字参数``plt.plot``指定了线条和标记的广泛属性:

````python
plt.plot(x, y, '-p', color='gray',
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.ylim(-1.2, 1.2);
````

这种灵活性在``plt.plot``函数中允许多种可能的可视化选项。

## 用``plt.scatter``来绘制散点图

第二种更有效的方法是用``plt.scatter``函数来创建散点图，它的使用方法和``plt.plot``非常类似。

````python
plt.scatter(x, y, marker='o');
````

``plt.scatter``于``plt.plot``的主要区别是，它可以用来创建散点图，其中每个点的属性(大小、颜色、边缘颜色等)可以单独控制或映射到数据。

让我们通过创建一个带有许多颜色和大小的点的随机散点图来展示这一点。

为了更好地看到重叠的结果，我们还将使用``alpha``关键字来调整透明度级别:

````python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar();  # show color scale
````

注意，颜色参数会自动映射到一个颜色表(这里显示的是``colorbar()``命令)，并且大小参数以像素形式给出。

通过这种方式，点的颜色和大小可以用来在可视化中传递信息，以便可视化多维数据。

例如，我们可以使用scikitt - learn的Iris数据，其中每个样本是三种类型的花，它们的花瓣和萼片的大小是经过仔细测量的:

````python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
````

我们可以看到这个散点图给了我们同时探索数据的四个不同维度的能力:

每个点的(x, y)位置对应于萼片的长度和宽度，点的大小与花瓣的宽度有关，颜色与花的特定种类有关。

像这样的多色和多特征散点图对数据的探测和显示都很有用。

## ``plot`` 对 ``scatter``: 效率

除了``plt.plot`` 和 ``plt.scatter``的不同特征，为什么你可以选择用一个来代替另一个?虽然对于少量的数据来说这并不重要，因为数据集要比几千个点更大，``plt.plot``'可以明显地比'``plt.scatter``更有效率。原因是``plt.scatter``有能力为每个点呈现不同的大小和/或颜色，因此渲染器必须单独完成每一点的构造。在另一方面``plt.plot``，这些点本质上都是彼此的克隆，因此确定这些点的外观的工作只对整个数据集进行一次。对于大型数据集，这两者之间的差异会导致截然不同的性能，因此，``plt.plot``在大型数据集中应该优于``plt.scatter`` 。
