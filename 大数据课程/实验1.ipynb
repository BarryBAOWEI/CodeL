# Matplotlib使用详解I 
##  简单的线形图

可能最简单的图是一个函数$y = f(x)$的可视化。

在这里，我们将首先考虑创建这种类型的简单图。

与下面的所有部分一样，我们将首先设置notebook以绘制和导入我们将要使用的packages:

````python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
````

对于所有的Matplotlib图，我们首先创建一个图形和一个坐标轴。

在最简单的形式中，可以创建一个图形和坐标轴如下:

````python
fig = plt.figure()
ax = plt.axes()
````

在Matplotlib中，*图*(类``plt.Figure``的一个实例)可以被认为是一个包含所有表示轴、图形、文本和标签的对象的容器。

我们在上面看到的是*坐标轴*(类``plt.Axes``的一个实例):一个带有刻度和标签的边界框，它最终将包含构成我们的可视化的plot元素。

在本次课程中，我们通常会使用变量名``fig``来引用一个图形实例，而 ``ax``则是指一个坐标轴实例或一组坐标轴实例。

一旦我们创建了一个坐标轴，我们就可以使用``ax.plot``函数来绘制一些数据。让我们从一个简单的正弦函数开始:

````python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
````

或者，我们可以使用pylab接口，并在后台为我们创建图和轴。

````python
plt.plot(x, np.sin(x));
````

如果我们想要创建一个多行的图形，我们可以简单地调用plot 函数多次:

````python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
````

这就是在Matplotlib中绘制简单函数的全部内容!

现在我们将深入了解如何控制轴和线的外观。

## 调整图形:线条的颜色和风格

你首先想要调整的可能是控制一个图形的线条颜色和风格。
``plt.plot()``函数接受额外的参数，可以用来指定这些参数。

为了调整颜色，你可以使用``color``这个词，它接受一个代表几乎任何可以想象到的颜色的字符串参数。

颜色可以通过多种方式来指定:

````python
plt.plot(x, np.sin(x - 0), color='blue')        # 通过名字指定颜色
plt.plot(x, np.sin(x - 1), color='g')           # 简称 (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # 0和1之间的灰度
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex 代码 (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # 取值从0到1的RGB tuple
plt.plot(x, np.sin(x - 5), color='chartreuse'); # 支持的所有HTML颜色名称
````

如果没有指定颜色，Matplotlib将自动循环通过一组默认颜色来进行多行。

同样的，线条风格也可以通过使用``linestyle`` 关键字来调整：

````python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# 为了简单，您可以使用以下代码:
plt.plot(x, x + 4, linestyle='-')  # 实线
plt.plot(x, x + 5, linestyle='--') # 虚线
plt.plot(x, x + 6, linestyle='-.') # 虚线、点线
plt.plot(x, x + 7, linestyle=':');  # 点线
````

如果你想要非常简洁，这些 ``linestyle``和``color``的代码可以组合成一个非关键字的``plt.plot()``功能:

````python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
````

这些单个字符的颜色代码反映标准缩写的RGB(红/绿/蓝)和CMYK颜色(青色/红色/黄色/黑色)系统中,常用的数码彩色图形。

还有许多其他的关键字参数可以用来调整情节的外观;有关更多细节，我建议您使用IPython的帮助工具查看``plt.plot()`` 函数的使用。

## 调整图形:轴的极限

Matplotlib在为你的图形选择默认的坐标轴上做的很好，但是有时候更好的控制是很好的。

最基本的调整轴限的方法是使用``plt.xlim()`` 和 ``plt.ylim()`` 的方法:

````python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
````

如果出于某种原因，您想要反向显示任意一个轴，您可以简单地颠倒参数的顺序:

````python
plt.plot(x, np.sin(x))

plt.xlim(10, 0)
plt.ylim(1.2, -1.2);
````

一个有用的相关方法是 ``plt.axis()``，允许您通过一个指定``[xmin, xmax, ymin, ymax]``的列表来设置``x``和``y``的极限：

````python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]);
````

``plt.axis()``方法甚至超出了这个范围，允许你做一些事情，比如自动收紧当前图形的边界:

````python
plt.plot(x, np.sin(x))
plt.axis('tight');
````

它甚至允许更高级别的规范，比如保证一个相等的宽比，这样在屏幕上，``x``中的一个单位就等于``y``中的一个单位:

````python
plt.plot(x, np.sin(x))
plt.axis('equal');
````

## 图形的标签

作为本实验的最后一部分，我们将简要地看一下图形的标签:标题、轴标签和简单的说明。

标题和轴标签是最简单的标签，有一些方法可以用来快速设置它们:

````python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
````

这些标签的位置、大小和样式可以使用可选参数来调整。

当在单个轴中显示多个行时，创建一个标记每行类型的图形说明是很有用的。

同样，Matplotlib有一种快速创建这样一个说明的内置方法。

这是通过``plt.legend()`` 的方法来完成的。

虽然有几种有效的使用方法，但我发现最容易的方法是使用``label``的plot函数关键字来指定每行的标签:

````python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend();
````

正如您所看到的，``plt.legend()``函数可以跟踪线条样式和颜色，并与正确的标签匹配。

## 除此之外: Matplotlib陷阱

虽然大多数 ``plt``函数可以直接转化为 ``ax``的方法(如``plt.plot()`` → ``ax.plot()``, ``plt.legend()`` → ``ax.legend()``等)，但不是对于所有的命令都是这样的。

特别地，设置坐标轴极限、标签和标题的函数被稍微修改了。

要在matlabstyle函数和面向对象的方法之间进行转换，请进行以下更改:

- ``plt.xlabel()``  → ``ax.set_xlabel()``
- ``plt.ylabel()`` → ``ax.set_ylabel()``
- ``plt.xlim()``  → ``ax.set_xlim()``
- ``plt.ylim()`` → ``ax.set_ylim()``
- ``plt.title()`` → ``ax.set_title()``

在面向对象的界面中绘图，而不是单独调用这些函数，使用``ax.set()``方法来设置所有这些属性通常更方便:

````python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot');
````
