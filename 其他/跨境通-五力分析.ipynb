## 跨境电商行业

分为进口电商与出口电商，又进一步有2C模式和2B模式，我们以2C为主要讨论对象

## 进口跨境电商的波特五力分析

### 同业的竞争

国内的进口市场基本被网易考拉、小红书等进口跨境电商与天猫国际、京东海囤全球等电商龙头的全球业务所占据

跨境通主打母婴用品与美容彩妆，以自建的五洲会与收购的优1宝贝作为拓展国内市场的重要途径

### 新进入者的威胁

跨境进口电商平台的技术壁垒非常低、新进入者无数

### 替代品的威胁

MADE IN CHINA全面碾压时间，这有可能吗？没有

### 上游溢价能力

（C2C）个人、海外代购：毫无议价能力

代理商、供货商、仓储：随规模提升能有一定的议价能力

### 下游议价能力

平台相互价格竞争明显，价格基本没有空间

## 出口跨境电商的波特五力分析

### 同业的竞争

主要与eBay、亚马逊、wish等国际上现有的开放平台以及环球易购、米兰网等国内自营平台开展竞争

跨境通旗下的环球易购拥有Gearbest、Sammydress等自建专业平台，是我国出口跨境电商的主要龙头

国内的出口跨境电商更多的以服装作为主要业务，环球易购的Sammydress、兰亭集势、米兰网都以服装作为主要产品，跨境通还自有知名服装品牌ZAFUL与首饰品牌LANGRIA

跨境通在此基础上还有3C类自营平台Gearbest和3C自营品牌Mpow，Mpow也是Amazon上的优质账号，曾有网红产品一说

### 新进入者的威胁

虽然跨境电商平台本身的技术壁垒非常低，但是跨境出口电商的仓储成本非常高，需要依赖于大量的规划优化，也具备规模经济优势的特点，同时海外品牌也对平台会有要求，新进入者也难以拿到授权。因此新进入者很难具被足够的竞争力

### 替代品的威胁

中国出口在性价比、质量等方面失去竞争力，这应该更不可能了

### 上游溢价能力

代理商、供货商、仓储：随规模提升能有一定的议价能力

自有品牌的优势，有极强的自主定价权

### 下游议价能力

同样有极强的比价空间

同样，自有品牌能够通过抬高在其他平台的价格，使得自己的平台具备极强的优势
