```python
import pandas as pd
import pypinyin


contentList = open('txt_sy.txt',encoding='gb2312').read().replace('\n','').split("、")
contentDf = pd.DataFrame({'目录':contentList,'拼音':[pypinyin.lazy_pinyin(i)[0].lower() for i in contentList]})
contentDfSort = contentDf.sort_values('拼音')
contentDfSort.to_csv('content.csv',encoding='gb2312')
contentDfSort['目录'].tolist()
```
