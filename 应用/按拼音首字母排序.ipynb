```python
from xpinyin import Pinyin
 
def pinyin_function(lis):             #输入一个名字的列表
    pin=Pinyin()
    result=[]
    for item in lis:
        if '\u4e00' <= item[0] <= '\u9fff':
            result.append((pin.get_pinyin(item),item))
        elif item[0].isalpha():
            result.append((item.lower(),item))
        else:
            result.append((item,item))
    result.sort()
    result_=[i[1] for  i in result]
    result_='\n'.join(result_)       #将排好序的结果连接，方便输出
    print(result_)                 #输出结果
    return result_
```
